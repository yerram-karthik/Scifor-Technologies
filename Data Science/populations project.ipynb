```python
#visualization [Heatmap]
numeric_data = data.select_dtypes(include=['int64', 'float64'])
correlation_matrix = numeric_data.corr()
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Heatmap of Numeric Variables')
plt.show()
```

```python
column_of_interest = '2022 Population'
plt.figure(figsize=(8, 6))
sns.boxplot(x=data[column_of_interest])
plt.title('Boxplot of {}'.format(column_of_interest))
plt.xlabel(column_of_interest)
plt.show()
```

```python
#Barplot
plt.figure(figsize=(10, 6))
sns.barplot(x='Continent', y='2022 Population', data=data)
plt.title('Population by Continent')
plt.xlabel('Continent')
plt.ylabel('Population')
plt.xticks(rotation=45)
plt.show()
```

```python
#Histogram
plt.figure(figsize=(10, 6))
plt.hist(data['2022 Population'], bins=20, color='skyblue', edgecolor='black')
plt.title('Histogram of 2022 Population')
plt.xlabel('Population')
plt.ylabel('Frequency')
plt.show()
```

```python
#Scatterplot

plt.figure(figsize=(10, 6))
plt.scatter(data['Capital'], data['2022 Population'], alpha=0.7, color='orange')
plt.title('Scatter Plot: GDP per Capita vs. 2022 Population')
plt.xlabel('Capital')
plt.ylabel('2022 Population')
plt.show()
```

```python
data = pd.DataFrame({
    'Country/Territory': ['China', 'India', 'United States', 'Indonesia', 'Pakistan'],
    'World Population Percentage': [17.88, 17.77, 4.24, 3.45, 2.96]
})

top_countries = data.nlargest(3, 'World Population Percentage')
other_percentage = 100 - top_countries['World Population Percentage'].sum()
other_countries = pd.DataFrame({'Country/Territory': ['Others'], 'World Population Percentage': [other_percentage]})
combined_data = pd.concat([top_countries, other_countries])
explode = [0.1 if country == 'China' else 0 for country in combined_data['Country/Territory']]

plt.figure(figsize=(12, 8))
plt.pie(combined_data['World Population Percentage'], labels=combined_data['Country/Territory'], autopct='%1.1f%%', startangle=90, explode=explode)
plt.title('World Population Percentage by Country')
plt.show()
```
