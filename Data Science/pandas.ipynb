```python
#create 20 students data in a dictionary and take input from the user and display the result

studentdata={"name":["karthik","yerram","nikhil","vedanta","vignesh","vihas","koushik","yaseen","shaik","khaitan","imtiaz","gowtham","hari","charan","vamsi","haripriya","sindhu","ruthika","prerana","parimi"],"age":[23,22,21,2,22,3,4,3,33,4,4,4,33,2,4,44,3,65,43,29],"address":["hyd","hyd","hyd","hyd","hyd","hyd","hyd","hyd","hyd","hyd","hyd","hyd","hyd","hyd","hyd","hyd","hyd","hyd","hyd","hyd"]}
search_name = input("Enter the name of the student you want to search for: ")

# Display the result
if search_name in studentdata['name']:
    index = studentdata['name'].index(search_name)
    student_info = {
        'name': studentdata['name'][index],
        'age': studentdata['age'][index],
        'address': studentdata['address'][index]
    }
    print(f"Student found!\nName: {student_info['name']}, Age: {student_info['age']}, Address: {student_info['address']}")
else:
    print("Student not found.")
```
