```python
class Patient:
    def __init__(self, patient_id, name, disease, doctor):
        self.patient_id = patient_id
        self.name = name
        self.disease = disease
        self.doctor = doctor

class RainbowHospital:
    def __init__(self):
        self.patients = []

    def admit_patient(self, patient):
        self.patients.append(patient)
        print("Patient admitted successfully!")

    def get_patient(self, key, value):
        found = False
        for patient in self.patients:
            if getattr(patient, key) == value:
                print(f"Patient found - ID: {patient.patient_id}, Name: {patient.name}, Disease: {patient.disease}, Doctor: {patient.doctor}")
                found = True
        if not found:
            print("Patient not found.")

    def show_all_patients(self):
        if self.patients:
            print("List of Patients in the Hospital:")
            for patient in self.patients:
                print(f"ID: {patient.patient_id}, Name: {patient.name}, Disease: {patient.disease}, Doctor: {patient.doctor}")
        else:
            print("No patients in the hospital.")

    def discharge_patient(self, key, value):
        for patient in self.patients:
            if getattr(patient, key) == value:
                self.patients.remove(patient)
                print("Patient discharged successfully!")
                return
        print("Patient not found.")

def main():
    hospital = RainbowHospital()

    while True:
        print("\nOptions:")
        print("1. Admit a new patient")
        print("2. Get patient details")
        print("3. Show all patients")
        print("4. Discharge a patient")
        print("5. Exit")

        choice = input("Enter your choice: ")

        if choice == '1':
            patient_id = input("Enter patient ID: ")
            name = input("Enter patient name: ")
            disease = input("Enter disease: ")
            doctor = input("Enter doctor in charge: ")
            patient = Patient(patient_id, name, disease, doctor)
            hospital.admit_patient(patient)
        elif choice == '2':
            key = input("Enter search key (ID/Name/Disease/Doctor): ")
            value = input("Enter search value: ")
            hospital.get_patient(key.lower(), value)
        elif choice == '3':
            hospital.show_all_patients()
        elif choice == '4':
            key = input("Enter discharge key (ID/Name/Disease/Doctor): ")
            value = input("Enter discharge value: ")
            hospital.discharge_patient(key.lower(), value)
        elif choice == '5':
            print("Exiting the program.")
            break
        else:
            print("Invalid choice. Please try again.")

if __name__ == "__main__":
    main()
```
