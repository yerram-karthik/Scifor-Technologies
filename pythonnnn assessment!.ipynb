```python
import os

def count_lines_words_letters(filename):
    total_lines = 0
    total_words = 0
    total_letters = 0
    line_start_count = {}

    with open(filename, 'r') as file:
        for line in file:
            total_lines += 1
            total_words += len(line.split())  
            total_letters += len(line) - line.count(' ')  

            first_letter = line[0].lower()
            if first_letter.isalpha():  
                line_start_count[first_letter] = line_start_count.get(first_letter, 0) + 1

    print("Total number of lines are :", total_lines)
    print("Total number of words are :", total_words)
    print("Total number of letters are :", total_letters)
    print("Lines starting with each letter are :")
    for letter, count in line_start_count.items():
        print(f"with {letter} are S: {count} lines")

file_path = r'C:\Users\yerra\OneDrive\Desktop\SCIFOR TRAINING\sample_file.txt'

with open(file_path, 'w') as file:
    file.write("Hello, world!\n")
    file.write("This is a sample file.\n")
    file.write("Python is awesome.\n")
    file.write("Let's count some lines.")


count_lines_words_letters(file_path)
```

```python
import tkinter as tk

class Person:
    def __init__(self, name, age):
        self._name = name
        self._age = age

    @property
    def name(self):
        return self._name

    @name.setter
    def name(self, value):
        self._name = value

    @property
    def age(self):
        return self._age

    @age.setter
    def age(self, value):
        if isinstance(value, int):
            self._age = value
        else:
            raise ValueError("Age must be an integer.")

    def __str__(self):
        return f"Name: {self.name}, Age: {self.age}"

def square_numbers(numbers):
    return map(lambda x: x**2, numbers)

def square_generator(numbers):
    for num in numbers:
        yield num**2

def display_result(result):
    for value in result:
        print(value)

person1 = Person("karthik", 22)
print(person1)
person1.age = 22
print(person1)

numbers = [1, 2, 3, 4, 5]
result = square_numbers(numbers)
display_result(result)

gen_result = square_generator(numbers)
display_result(gen_result)

class App:
    def __init__(self, master):
        self.master = master
        self.master.geometry("300x200")  
        self.label = tk.Label(master, text="Enter your name:")
        self.label.pack()
        self.entry = tk.Entry(master)
        self.entry.pack()
        self.button = tk.Button(master, text="Greet", command=self.greet)
        self.button.pack()

    def greet(self):
        name = self.entry.get()
        greeting = tk.Label(self.master, text=f"Hello, {name}!")
        greeting.pack()

root = tk.Tk()
app = App(root)
root.mainloop()
```

```python
import json

data = {
    "name": "Yerram Karthik",
    "age": 22,
    "address": {
        "city": "Hyderabad",
        "zipcode": "500070"
    },
    "interests": ["reading", "traveling", "coding"]
}

with open('data.json', 'w') as json_file:
    json.dump(data, json_file, indent=4)

with open('data.json', 'r') as json_file:
    loaded_data = json.load(json_file)

print("Loaded JSON data:")
print(loaded_data)
print()

print("Name:", loaded_data['name'])
print("Age:", loaded_data['age'])
print("City:", loaded_data['address']['city'])
print("Interests:", ', '.join(loaded_data['interests']))
print()

loaded_data['age'] = 22
loaded_data['address']['city'] = "Hyderabad"
loaded_data['interests'].append("cooking")

print("Modified JSON data:")
print(loaded_data)
print()

json_string = json.dumps(loaded_data, indent=4)
print("JSON string representation:")
print(json_string)
print()

with open('data.json', 'w') as json_file:
    json.dump(loaded_data, json_file, indent=4)

with open('data.json', 'r') as json_file:
    reloaded_data = json.load(json_file)

print("Reloaded JSON data after modification:")
print(reloaded_data)
```
